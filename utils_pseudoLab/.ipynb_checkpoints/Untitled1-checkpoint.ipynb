```python
%pylab inline
import numpy as np
```

```python
def linear_alpha(ep_init, ep_end, total_epochs, current_epoch):
    epochs = list(range(total_epochs))
    start_ep = ep_init
    end_ep = ep_end
    alpha = []

    init_val = 1
    end_val = 32

    alpha_mid = list(np.linspace(init_val, end_val, len(epochs[start_ep:end_ep])))
    alpha_init = [init_val for i in range(start_ep)]
    alpha_end = [end_val for i in range(len(epochs) - end_ep)]

    alpha.extend(alpha_init)
    alpha.extend(alpha_mid)
    alpha.extend(alpha_end)

    return alpha[current_epoch-1]
```

```python
alpha = []
nb_epochs = 400
epochs = list(range(nb_epochs))

for i in range(1,nb_epochs+1):
#     print(i, tunable_sigmoid_alpha(1, 32, nb_epochs, i, k))
    alpha.append(linear_alpha(1, 400, nb_epochs, i))
    

plot(epochs, alpha)
print(alpha[0], alpha[-1])
```

```python
def tunable_sigmoid_alpha(init_val, end_val, total_epochs, current_epoch, k):
    # blog here:
    # https://dhemery.github.io/DHE-Modules/technical/sigmoid/
    total_epochs = total_epochs -1
    current_epoch = current_epoch -1
    scaled_current_epoch = (current_epoch-(total_epochs/2))/(total_epochs/2)

    sig_num = scaled_current_epoch - k*scaled_current_epoch
    sigm_denom = k - (2*k*np.abs(scaled_current_epoch)) + 1

    sigm = sig_num/np.max([sigm_denom, 1e-6])
    val_range = end_val - init_val
    sigm_scaled = ((val_range/2)*sigm)+(val_range/2)

    return init_val+sigm_scaled
```

```python
alpha = []
nb_epochs = 400
epochs = list(range(nb_epochs))
k = -0.5

for i in range(1,nb_epochs+1):
#     print(i, tunable_sigmoid_alpha(1, 32, nb_epochs, i, k))
    alpha.append(tunable_sigmoid_alpha(1, 32, nb_epochs, i, k))
    

plot(epochs, alpha)
print(alpha[0], alpha[-1])
```

```python
def linear_alpha2(init_val, end_val, total_epochs, current_epoch):
    return list(np.linspace(init_val, end_val, total_epochs))[current_epoch-1]
```

```python
alpha = []
nb_epochs = 400
epochs = list(range(nb_epochs))
k = 0.0

for i in range(1,nb_epochs+1):
#     print(i, tunable_sigmoid_alpha(1, 32, nb_epochs, i, k))
    alpha.append(linear_alpha2(0.4, 0.1, nb_epochs, i))
    

plot(epochs, alpha)
print(alpha[0], alpha[-1])
```

```python
def exponential_fx(init_val, end_val, total_epochs, current_epoch, k):
#     total_epochs = total_epochs -1
#     current_epoch = current_epoch -1
    
    b = (np.log(end_val) - (total_epochs*np.log(init_val)))/(-total_epochs + 1)
    a = np.log(init_val) - b

    return np.exp((a*current_epoch) + b)
```

```python
alpha = []
nb_epochs = 400
epochs = list(range(nb_epochs))
k = 0.0

for i in range(1,nb_epochs+1):
#     print(i, tunable_sigmoid_alpha(1, 32, nb_epochs, i, k))
    alpha.append(exponential_fx(0.1, 8, nb_epochs, i, k))
    

plot(epochs, alpha)
print(alpha[0], alpha[-1])
```

```python
def log_fx(init_val, end_val, total_epochs, current_epoch, k):
#     total_epochs = total_epochs -1
#     current_epoch = current_epoch -1
    
    b = (np.exp(end_val) - (total_epochs*np.exp(init_val)))/(-total_epochs + 1)
    a = np.exp(init_val) - b

    return np.log((a*current_epoch) + b)
```

```python
alpha = []
nb_epochs = 400
epochs = list(range(nb_epochs))
k = 0.0

for i in range(1,nb_epochs+1):
#     print(i, tunable_sigmoid_alpha(1, 32, nb_epochs, i, k))
    alpha.append(log_fx(0.1, 8, nb_epochs, i, k))
    

plot(epochs, alpha)
print(alpha[0], alpha[-1])
```
